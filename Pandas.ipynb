Manipulating a Dataframe

```python
# adding a column to a dataframe
california_df['Price'] = california_dataset.target
```

```python
california_df.head()
```

```python
# removing a row
california_df.drop(index = 1, axis = 0) # 1st index row i.e, 2nd row will be deleted (cauyse 0 based indexing)
```
